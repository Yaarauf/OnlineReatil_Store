```python
# Filter outliers based on numerical features
numerical_cols = retaildf.select_dtypes(include=np.number).columns
numerical_cols = numerical_cols.drop(['CustomerID', 'InvoiceMonth'])  # Excluding CustomerID and InvoiceMonth

# Calculate Q1, Q3, and IQR only for the selected numerical columns
Q1 = retaildf[numerical_cols].quantile(0.25)
Q3 = retaildf[numerical_cols].quantile(0.75)
IQR = Q3 - Q1

# Calculate lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter outliers for the selected numerical columns
retail_filtered = retaildf[~((retaildf[numerical_cols] < lower_bound) | (retaildf[numerical_cols] > upper_bound)).any(axis=1)]
```
